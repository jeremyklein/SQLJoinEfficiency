```python
import pymssql
import json

with open("config.json", "r") as config:
    config_dict = json.load(config)
    
```

```python
database_name = config_dict["database_name"]
database_server = config_dict["server"]
DB_user = config_dict["DB_user"]
DB_password = config_dict["DB_password"]

cnxn = pymssql.connect(database_server, DB_user,DB_password, database_name)
```

```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine("mssql+pymssql://"+DB_user+":"+DB_password+"@"+database_server+"/"+database_name)

def csv_to_sql(csv_file_name, table_name):
    dataframe = pd.read_csv(csv_file_name)
    dataframe.to_sql(table_name,engine,index=False)
```

```python
import os

sample_data_directory = "sample_data/"

files = os.listdir(sample_data_directory)
for file in files:
    file_path = sample_data_directory + file
    table_name = file.split(".")[0]
    csv_to_sql(file_path, table_name)
```

```python
cursor = cnxn.cursor()
cursor.execute("""
SET showplan_xml ON;
""")

cursor.execute("""
Select *
FROM patients p
LEFT JOIN  observations o
      on o.PATIENT = p.ID
WHERE
      o.PATIENT is NULL
      AND o.CODE IS NULL 
      AND o.DATE IS NULL
      AND o.DESCRIPTION IS NULL
      AND o.ENCOUNTER IS NULL
      AND o.TYPE IS NULL
      AND o.UNITS IS NULL
      AND o.VALUE IS NULL;
""")
```
